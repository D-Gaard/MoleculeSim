```python
import matplotlib.pyplot as plt
```

```python
fig_size = 10
lim_size = 8
radius = 1

plt.figure(figsize=[fig_size, fig_size])

plt.axes([0.0, 0.0, 1, 1], xlim=(0, lim_size), ylim=(0, lim_size))


points_whole_ax = fig_size * 1 * 72    # 1 point = dpi / 72 pixels
points_radius = 2 * radius / lim_size * points_whole_ax
plt.scatter(3, 3, s=points_radius**2, color='r')
plt.grid()
plt.show()
```
