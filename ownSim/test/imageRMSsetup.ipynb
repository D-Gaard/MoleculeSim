```python
def duplicate_image(img,dupX,dupY):
  height,width = img.shape
  new_height = height*dupY
  new_width = width*dupX
  duplicated_image = np.zeros((new_height, new_width))

  for i in tqdm(range(dupX)):
    for j in range(dupY):
      paste_y = j * height
      paste_x = i * width

      duplicated_image[paste_y:paste_y + height, paste_x:paste_x + width] = img

  return duplicated_image
```

```python
orig_size = np.array([8000,8000])
orig_img = np.ones(orig_size)
orig_img[2000:4001,2000:4001] = 200

plt.subplot(1,3,1)
plt.imshow(orig_img)

plt.subplot(1,3,2)
_dup = duplicate_image(orig_img,3,4)
plt.imshow(_dup)

plt.subplot(1,3,3)
_dup_resize = resize(_dup,orig_size)
plt.imshow(_dup_resize)
plt.show()
```
